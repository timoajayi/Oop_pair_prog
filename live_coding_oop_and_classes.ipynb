# 1. Object Oriented Programming (OOP)

**Object-Oriented Programming (OOP)** is a programming paradigm based on the concept of **objects**:
> Programs are designed by creating objects that interact with each other


## What are Objects?

Objects are the building blocks of OOP. Think of them as digital representations of real-world things:

1. **Objects represent real-world entities** (tangible or conceptual)
   - Examples: a panda, a car, a bank account, a user profile

2. **Objects contain data** (called _attributes_ or _properties_ in Python)
   - Attributes describe the **state** of an object
   - Example: A panda has a `name`, `energy` level, and `bamboo_eaten` count

3. **Objects contain functions** (called _methods_ in Python)
   - Methods define **behaviors** that can change the object's state or perform actions
   - Example: A panda can `eat()` bamboo or `nap()` to restore energy

### Designing a Panda as an Object

If we think of a **Panda** as an object, what properties and behaviors would it have?

**Attributes (Data):**
- `name` - What is the panda called?
- `energy` - How energetic is the panda? (0-100)
- `bamboo_eaten` - How much bamboo has it consumed?

**Methods (Behaviors):**
- `eat()` - Panda eats bamboo and gains energy
- `nap()` - Panda sleeps and restores energy  
- `play()` - Panda plays and uses energy

# 2. Classes

A **class** defines the data formats (attributes) and available procedures (methods) for a given type of object.

## Classes vs. Objects

Think of a class as a **blueprint** or **template**:
- The **Panda class** is the blueprint that defines what all pandas have in common
- Each **panda object** (like "Bao Bao" or "Mei Xiang") is a specific instance created from that blueprint
