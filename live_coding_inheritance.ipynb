# Inheritance in Object-Oriented Programming


**Inheritance** is one of the fundamental pillars of Object-Oriented Programming. It allows you to create hierarchical relationships between classes, where **child classes** (also known as **subclasses** or **derived classes**) inherit attributes and methods from their **parent classes** (also called **superclasses** or **base classes**).

## Why Use Inheritance?

1. **Code Reusability** - Write once, use many times! No need to duplicate code.
2. **Maintainability** - Fix bugs in one place, and all child classes benefit.
3. **Extensibility** - Easily add new features without modifying existing code.
4. **Natural Modeling** - Represents real-world "is-a" relationships.

## Python Syntax for Inheritance

Creating a child class in Python is simple:

```python
# Parent class
class Panda:
    pass

# Child class inherits from Panda
class KungFuPanda(Panda):
    pass
```

**Key points:**
- Put the parent class name in parentheses: `class Child(Parent):`
- Use `super()` to access parent class methods
- Child automatically gets all parent attributes and methods

Let's see this in action! 


````python
# Our Lovely Panda
class Panda:
    """Base class for a Panda"""
    def __init__(self, name, energy=50):
        self.name = name
        self.energy = energy
        self.bamboo_eaten = 0
        print(f"ğŸ¼ {self.name} the Panda is created!")
        
    def __repr__(self):
        return f"Panda(name={self.name}, energy={self.energy}, bamboo_eaten={self.bamboo_eaten})"
    
    def eat(self, bamboo_amount=1):
        """Panda eats bamboo and gains energy"""
        self.bamboo_eaten = self.bamboo_eaten +  bamboo_amount
        self.energy = self.energy +  bamboo_amount * 10
        print(f"{self.name} ate {bamboo_amount} bamboo. Energy: {self.energy}")
        
    def nap(self, duration=1):
        """Panda takes a nap and gains energy"""
        self.energy = self.energy +  duration * 20
        print(f"{self.name} napped for {duration} hour(s). Energy: {self.energy}")
````
